```python
newlst = []

for elem in lst:
    file = open('files/'+elem, 'r')
    text = file.read()
    beginning = text[:1000]
    end = text[-2000:]
    text = beginning + end
    
    date = re.search(r'[0-9]*(.) [a-zA-Zéèà]* 201[0-9]', text).group()
    newlst.append(date)
```

```python
newlst
```

```python
pd.DataFrame(newlst).to_csv('dates.csv')
```
